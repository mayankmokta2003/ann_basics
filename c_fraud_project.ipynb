```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("credit_fraud.csv")
```

```python

df = df.drop("age", axis=1)
df.isna().sum()
```

```python
ta_mean = df["transaction_amount"].mean()
df["transaction_amount"] = df["transaction_amount"].fillna(ta_mean)
```

```python
ta_mean = df["account_balance"].mean()
df["account_balance"] = df["account_balance"].fillna(ta_mean)
```

```python
ta_mean = df["num_transactions_today"].mean()
df["num_transactions_today"] = df["num_transactions_today"].fillna(ta_mean)
```

```python
ta_mean = df["is_foreign_transaction"].mean()
df["is_foreign_transaction"] = df["is_foreign_transaction"].fillna(ta_mean)
```

```python
ta_mean = df["transaction_hour"].mean()
df["transaction_hour"] = df["transaction_hour"].fillna(ta_mean)
```

```python
ta_mean = df["prev_fraud_flag"].mean()
df["prev_fraud_flag"] = df["prev_fraud_flag"].fillna(ta_mean)
```

```python
ta_mean = df["merchant_distance_km"].mean()
df["merchant_distance_km"] = df["merchant_distance_km"].fillna(ta_mean)
```

```python
ta_mean = df["merchant_risk_score"].mean()
df["merchant_risk_score"] = df["merchant_risk_score"].fillna(ta_mean)
```

```python
df = df.dropna()
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop("is_fraud", axis=1)
y = df["is_fraud"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras import layers

model = Sequential([
    layers.Dense(16, input_dim = 8, activation="relu"),
    layers.Dense(8, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy",metrics=["accuracy"])

history = model.fit(
    X_train_scaled, y_train,
    validation_data = (X_test_scaled, y_test),
    epochs = 100, 
    batch_size = 1028,
    verbose = 1
)
```

```python
loss, acc = model.evaluate(X_test_scaled, y_test, verbose = 1)
print(acc)
```
